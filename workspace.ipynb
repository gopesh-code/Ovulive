```python
import pandas as pd
import numpy as np
import pickle
```

```python
data = pd.read_csv("dataset/FedCycleData071012 (2)_fixed.csv",na_values=[' '])
data=data.fillna(method ='pad')
```

```python
data.head()
```

```python
X=data[['ReproductiveCategory','LengthofLutealPhase','LengthofMenses', 'MensesScoreDayOne', 'MensesScoreDayTwo', 'MensesScoreDayThree', 'MensesScoreDayFour', 'MensesScoreDayFive',  'MensesScoreDaySix', 'NumberofDaysofIntercourse', 'IntercourseInFertileWindow', 'UnusualBleeding']]
Y=data['LengthofCycle']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.pipeline import make_pipeline
```

```python
def create_train_model(scaler, regressor, x_train, y_train):
    model = make_pipeline(scaler, regressor)
    model.fit(x_train, y_train)
    return model
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, median_absolute_error, explained_variance_score, max_error, mean_poisson_deviance, mean_gamma_deviance
```

```python
def show_model_results(model, x_test, y_test):
    y_pred = model.predict(x_test)
    print("Mean Squared Error: ", mean_squared_error(y_test, y_pred))
    print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
    print("Median Absolute Error: ", median_absolute_error(y_test, y_pred))
    print("Explained Variance Score: ", explained_variance_score(y_test, y_pred))
    print("Max Error: ", max_error(y_test, y_pred))
    print("Mean Poisson Deviance: ", mean_poisson_deviance(y_test, y_pred))
    print("Mean Gamma Deviance: ", mean_gamma_deviance(y_test, y_pred))
    print("R2 Score: ", r2_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, SGDRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
sgd_reg_model = create_train_model(StandardScaler(), SGDRegressor(), x_train, y_train)
show_model_results(sgd_reg_model, x_test, y_test)
```

```python
rndf_reg_model = create_train_model(StandardScaler(), RandomForestRegressor(), x_train, y_train)
show_model_results(rndf_reg_model, x_test, y_test)
```

```python
svr_reg_model = create_train_model(StandardScaler(), SVR(), x_train, y_train)
show_model_results(svr_reg_model, x_test, y_test)
```

```python
log_reg_model = create_train_model(StandardScaler(), LogisticRegression(), x_train, y_train)
show_model_results(log_reg_model, x_test, y_test)
```

```python
dec_tree_reg_model = create_train_model(StandardScaler(), DecisionTreeRegressor(), x_train, y_train)
show_model_results(dec_tree_reg_model, x_test, y_test)
```

```python
grad_boost_reg_model = create_train_model(StandardScaler(), GradientBoostingRegressor(), x_train, y_train)
show_model_results(grad_boost_reg_model, x_test, y_test)
```

```python
kmeans_reg_model = create_train_model(StandardScaler(), KMeans(), x_train, y_train)
show_model_results(kmeans_reg_model, x_test, y_test)
```

```python
agglom_reg_model = create_train_model(StandardScaler(), AgglomerativeClustering(), x_train, y_train)
show_model_results(agglom_reg_model, x_test, y_test)
```
